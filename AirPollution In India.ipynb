```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print('setup complete')
```

```python
dta = pd.read_excel('AirQuality.xlsx')
dta.head()
```

```python
dta=dta.drop(columns=['Country','lastupdate'])
```

```python
dta.head()
```

```python
dta.columns=dta.columns.str.lower()
```

```python
dta.head()
```

```python
dta['place'] = dta['place'].apply(lambda x: x.split(' ')[-1])
```
